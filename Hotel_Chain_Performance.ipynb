```python
df['Hotel_Type'] = df['Hotel_Type'].ffill()
```

```python
df['Event_Season'].fillna('Off season', inplace=True)
```

```python
df['Avg_Room_Price']= df['Avg_Room_Price'].ffill()
```

```python
df['Customer_Rating'].fillna(df['Customer_Rating'].median(), inplace=True)
```

```python
df['Cancellations'].fillna(0, inplace=True)
```

```python
df.describe(include ="all")
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
city_Daily_Revenue = df.groupby("City")["Daily_Revenue"].sum()
plt.figure(figsize=(12,4))
plt.bar(city_Daily_Revenue.index,city_Daily_Revenue.values,color = "blue",edgecolor = "black")
plt.title("Daily Revenue of citys")
plt.xlabel("City Names")
plt.ylabel("Total Daily Revenue in .rs")
plt.show()
```

```python
plt.figure(figsize=(8, 5))
plt.hist(df["Net_Profit"], bins=10, color="red", edgecolor="black") 
plt.title("Net Profit") 
plt.xlabel("Net Profit in Lakh")  
plt.ylabel("Frequency")  
plt.show()
```

- This graph is a **right skewed distribution**
- in this histogram the **net profit of 1800 hotels falls between 0 to 1 lakh**
- **This range has the highest number of hotels** indicating that **most hotels earned a lower net profit**
- A very small number of hotels have a **net profit between 4 to 7 lakh**

```python
Hotel_net_profite = df.groupby("Hotel_Type")["Net_Profit"].mean()
plt.bar(Hotel_net_profite.index, Hotel_net_profite.values, color="green", edgecolor="black")
plt.title("Average net profite by Hotel Type")
plt.xlabel("Hotel Type")
plt.ylabel("Average Customer Rating")
plt.show()
```

```python
Room_booked_Event_season = df.groupby("City")["Daily_Revenue"].sum()
plt.bar(Room_booked_Event_season.index, Room_booked_Event_season.values, color="green", edgecolor="black")
plt.xticks(rotation=90)
plt.xlabel("Hotel Type")
plt.ylabel("Average Customer Rating")
plt.title("Average Ratings by Hotel Type")
plt.show()
```

```python
plt.hist(df['Customer_Rating'], bins=5, color='orange', edgecolor='black')
plt.title('Customer Rating Distribution')
plt.xlabel('Rating')
plt.ylabel('Frequency')
plt.tight_layout()
plt.show()
```

```python
group_sum = df.groupby('Hotel_Type')['Profit_Margin_Percent'].sum()
plt.figure(figsize=(6, 6))
plt.pie(group_sum.values, labels=group_sum.index, autopct='%1.1f%%', startangle=90)
plt.title("Hotel Type vs Profit Margin percentage")
plt.show()
```

```python
plt.scatter(df['Occupancy_Rate'], df['Net_Profit'], alpha=0.6, color='purple',edgecolor = "black")
plt.title('Occupancy Rate vs Net Profit')
plt.xlabel('Occupancy Rate %')
plt.ylabel('Net Profit million')
plt.show()
```

```python
plt.boxplot(df['Profit_Margin_Percent'])
plt.title('Profit Margin Distribution')
plt.ylabel('Profit Margin')
plt.show()
```

```python
sns.violinplot(x='Event_Season', y='Net_Profit', data=df, palette='Dark2')
plt.title('Net Profit by Event Season')
plt.xlabel('Event Season')
plt.ylabel('Net Profit')
plt.xticks(rotation = 90)
plt.show()
```

```python
plt.figure(figsize = (10,7))
cols = ['Rooms_Available', 'Rooms_Booked', 'Occupancy_Rate', 'Avg_Room_Price','Daily_Revenue', 'Operating_Cost', 'Net_Profit','Profit_Margin_Percent',
        'Customer_Rating', 'Cancellations', 'Repeat_Customers_Percent']
corr_matrix = df[cols].corr()
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm')
plt.title('Correlation Heatmap of Hotel Metrics')
plt.show()
```

```python
sns.pairplot(corr_matrix, diag_kind='kde', corner=True)
plt.suptitle('Pair Plot of Hotel Booking Metrics')
plt.tight_layout()
plt.show()
```
