```python
df['BMI'].fillna( df['BMI'].median(),inplace = True)
df['Sleep_Hours'].fillna(df['Sleep_Hours'].median(), inplace=True)
df['Calories_Consumed'].fillna(df['Calories_Consumed'].median(), inplace=True)
```

```python
df['Gender'] = df['Gender'].str.title()
```

```python
#1. What is the distribution of gender in the dataset?  Pie Chart / Donut Chart
gender_counts = df['Gender'].value_counts()
plt.figure(figsize=(6, 6))
plt.pie(gender_counts, labels=gender_counts.index, autopct='%1.1f%%',startangle=90)
plt.title('Gender Distribution in Dataset')
plt.show()

# Donut chart version
plt.figure(figsize=(6, 6))
plt.pie(gender_counts, labels=gender_counts.index, autopct='%1.1f%%', startangle=90, wedgeprops=dict(width=0.4))
plt.title('Gender Distribution (Donut Style)')
plt.tight_layout()
plt.show()
```

```python
#2. How is age distributed across users?  Histogram / Bar Chart
plt.figure(figsize=(8, 5))
plt.hist(df['Age'], bins=10, color='skyblue', edgecolor='black')
plt.title('Age Distribution of Users')
plt.xlabel('Age')
plt.ylabel('Number of Users')
plt.tight_layout()
plt.show()

age_counts = df['Age'].value_counts()
plt.figure(figsize=(10, 5))
plt.bar(age_counts.index, age_counts.values, color='skyblue', edgecolor='black')
plt.title('User Count by Age')
plt.xlabel('Age')
plt.ylabel('Users')
plt.show()
```

```python
#3. What is the distribution of daily calorie intake?  Histogram / Box Plot
plt.figure(figsize=(8, 5))
plt.hist(df['Calories_Consumed'], bins=10, color='skyblue', edgecolor='black')
plt.title('daily calorie intake Distribution')
plt.xlabel('Calories Consumed')
plt.ylabel('Users')
plt.show()

#box plot
plt.figure(figsize=(6, 5))
plt.boxplot(df['Calories_Consumed'])
plt.title('Box Plot of Daily Calorie Intake')
plt.ylabel('Calories Consumed')
plt.show()
```

```python
#4. How do protein, carbs, and sugar grams vary among users?  Box Plot / Histogram
plt.figure(figsize=(10, 6))
plt.boxplot([df['Protein_Grams'], df['Carbs_Grams'], df['Sugar_Grams']],patch_artist=True)
plt.title('Distribution of Protein, Carbs, and Sugar Intake')
plt.ylabel('Grams')
plt.show()

fig, axs = plt.subplots(1, 3, figsize=(18, 5))  
axs[0].hist(df['Protein_Grams'], bins=10, color='red', edgecolor='black')
axs[0].set_title('Protein Intake Distribution')
axs[0].set_xlabel('Grams')
axs[0].set_ylabel('Users')

axs[1].hist(df['Carbs_Grams'], bins=10, color='green', edgecolor='black')
axs[1].set_title('Carbs Intake Distribution')
axs[1].set_xlabel('Grams')
axs[1].set_ylabel('Users')

axs[2].hist(df['Sugar_Grams'], bins=10, color='blue', edgecolor='black')
axs[2].set_title('Sugar Intake Distribution')
axs[2].set_xlabel('Grams')
axs[2].set_ylabel('Users')
plt.show()
```

```python
#5. What is the average sleep duration among users?  Bar Chart / Line Chart
sleep_by_weight = df.groupby('Weight_Change_Category')['Sleep_Hours'].mean()

plt.figure(figsize=(6, 4))
plt.bar(sleep_by_weight.index, sleep_by_weight.values, color='lightgreen', edgecolor='black')
plt.title('Average Sleep Duration by Weight Change Category')
plt.xlabel('Weight Change Category')
plt.ylabel('Average Sleep Hours')
plt.tight_layout()
plt.show()
```

```python
#6.How many minutes of exercise do users perform on average?  Histogram / Bar Chart

plt.figure(figsize=(8, 5))
plt.hist(df["Exercise_Minutes"], bins=10, color="blue", edgecolor="black")
plt.title("Distribution of Daily Exercise Minutes")
plt.xlabel("Exercise Minutes")
plt.ylabel("Users")
plt.show()

exercise_mean = df["Exercise_Minutes"].mean()
plt.figure(figsize=(6, 4))
plt.bar(["Average Exercise"], exercise_mean, color="skyblue", edgecolor="black")
plt.title("Average Daily Exercise Duration")
plt.ylabel("Minutes")
plt.tight_layout()
plt.show()
```

```python
#7. How is water intake (liters) distributed?  Histogram / Box Plot
plt.figure(figsize=(8, 5))
plt.hist(df["Water_Intake_Liters"], bins=10, color="blue", edgecolor="black")
plt.title("Distribution of Daily Exercise Minutes")
plt.xlabel("Exercise Minutes")
plt.ylabel("Users")
plt.show()

plt.figure(figsize=(6, 4))
plt.boxplot(df["Water_Intake_Liters"])
plt.title("Box Plot of Daily Water Intake")
plt.ylabel("Water Intake Liters")
plt.tight_layout()
plt.show()
```

```python
#8. What is the distribution of mood scores (110)?  Bar Chart / Histogram
mood_counts = df["Mood_Score"].value_counts()
plt.figure(figsize=(8, 5))
plt.bar(mood_counts.index,mood_counts.values,color='skyblue', edgecolor='black')
plt.title('Mood Score Distribution')
plt.xlabel('Mood Score 1 - 10')
plt.ylabel('Users')
plt.xticks(range(1,11))
plt.show()

plt.figure(figsize=(8, 5))
plt.hist(df['Mood_Score'], bins=10, color='purple', edgecolor='black')
plt.title('Histogram of Mood Scores')
plt.xlabel('Mood Score')
plt.ylabel('Frequency')
plt.show()
```

```python
df['Weight_Change_Category'] = df['Weight_Change_Category'].str.strip().str.title()
weight_counts = df['Weight_Change_Category'].value_counts()
plt.figure(figsize=(6, 6))
plt.pie(weight_counts, labels=weight_counts.index, autopct='%1.1f%%', startangle=90)
plt.title('User Distribution by Weight Change Category')
plt.show()

plt.figure(figsize=(6, 6))
plt.pie(weight_counts, labels=weight_counts.index, autopct='%1.1f%%', startangle=90,wedgeprops=dict(width=0.4))
plt.title('User Distribution by Weight Change Category')
plt.tight_layout()
plt.show()
```

```python
# 10. How does calorie consumption vary across weight change categories?  Box Plot / Bar Chart
plt.figure(figsize=(8, 5))
df['Weight_Change_Category'] = df['Weight_Change_Category'].str.title()
sns.boxplot(data =df,x='Weight_Change_Category',y='Calories_Consumed')
plt.title('Calorie Consumption by Weight Change Category')
plt.xlabel('Weight Change Category')
plt.ylabel('Calories Consumed')
plt.show()

calorie_means = df.groupby('Weight_Change_Category')['Calories_Consumed'].mean()
plt.figure(figsize=(6, 4))
plt.bar(calorie_means.index, calorie_means.values, color='skyblue', edgecolor='black')
plt.title('Avg Calorie Consumption by Weight Change Category')
plt.xlabel('Weight Change Category')
plt.ylabel('Avg Calorie Consumption')
plt.tight_layout()
plt.show()
```

```python
#11. How does exercise duration differ among Gain, Maintain, Lose groups?  Box Plot / Bar Chart
plt.figure(figsize=(8, 5))
sns.boxplot(data = df,x= 'Weight_Change_Category', y = 'Exercise_Minutes',palette = "tab20")
plt.title('Exercise Duration by Weight Change Category')
plt.xlabel('Weight Change Category')
plt.ylabel('Exercise Minutes')
plt.show()

exercise_means = df.groupby('Weight_Change_Category')['Exercise_Minutes'].mean()
plt.figure(figsize=(6, 4))
#plt.bar(df["Weight_Change_Category"], df["Exercise_Minutes"], color='skyblue', edgecolor='black')
plt.bar(exercise_means.index,exercise_means.values, color='skyblue', edgecolor='black')
plt.title('Exercise Duration by Weight Change Category')
plt.xlabel('Weight Change Category')
plt.ylabel('Exercise Minutes')
plt.show()
```

```python
#12. Does better sleep correspond to higher mood scores?  Scatter Plot / Line Chart
plt.figure(figsize=(8, 5))
plt.scatter(df['Sleep_Hours'], df['Mood_Score'],color='blue', edgecolor='black')
plt.title('Relationship Between Sleep Duration and Mood Score')
plt.xlabel('Sleep Hours')
plt.ylabel('Mood Score')
plt.show()

sleep_mood = df.groupby('Sleep_Hours')['Mood_Score'].mean()

plt.figure(figsize=(8, 5))
plt.plot(sleep_mood.index, sleep_mood.values, marker='o', linestyle='-', color='green')
plt.title('Average Mood Score by Sleep Duration')
plt.xlabel('Sleep Hours')
plt.ylabel('Average Mood Score')
plt.show()
```

```python
#13. What is the relationship between calories consumed and exercise minutes?  Scatter Plot / Pairplot
plt.figure(figsize=(8, 5))
plt.scatter(df['Calories_Consumed'], df['Exercise_Minutes'], color='blue', edgecolor='black')
plt.title('Calories vs Exercise Duration')
plt.xlabel('Calories Consumed')
plt.ylabel('Exercise Minutes')
plt.show()

sns.pairplot(df[['Calories_Consumed', 'Exercise_Minutes']])
plt.suptitle('Pairplot of  Calories Consumed and  Exercise Minutes')
plt.show()
```

```python
#14. Do users who consume more protein tend to lose or gain weight?  Box Plot / Bar Chart
plt.figure(figsize=(8, 5))
sns.boxplot(data = df,x='Weight_Change_Category', y='Protein_Grams',palette='tab10')
plt.title('Protein Intake by Weight Change Category')
plt.xlabel('Weight Change Category')
plt.ylabel('Protein in Grams')
plt.show()

protein_means = df.groupby('Weight_Change_Category')['Protein_Grams'].mean()
plt.figure(figsize=(6, 4))
plt.bar(protein_means.index, protein_means.values, color='skyblue', edgecolor='black')
plt.title('Avg Protein Intake by Weight Change Category')
plt.xlabel('Weight Change Category')
plt.ylabel('Avg protein in Grams')
plt.show()
```

```python
#15. Is there a relationship between sugar intake and BMI?  Scatter Plot / Line Chart
plt.scatter(df["Sugar_Grams"],df["BMI"],color = "red",edgecolor = "black")
plt.title("Sugar intake vs BMI")
plt.xlabel("suger Grams")
plt.ylabel("BMI")
plt.show()

sugar_BMI = df.groupby("Sugar_Grams")["BMI"].mean()
plt.plot(sugar_BMI.index,sugar_BMI.values, marker = "o",color = "green")
plt.title('Avg BMI by Sugar Intake')
plt.xlabel('Sugar Intake Grams')
plt.ylabel('Avg BMI')
plt.show()
```

```python
# 16. Does hydration level impact weight change outcome?  Box Plot / Bar Chart
sns.boxplot(data=df,x="Weight_Change_Category",y="Water_Intake_Liters")
plt.title('Hydration Level by Weight Change Category')
plt.xlabel('Weight Change Category')
plt.ylabel('Water Intake Liters')
plt.show()

#water_means = df.groupby('Weight_Change_Category')['Water_Intake_Liters'].mean()
plt.figure(figsize=(6, 4))
#plt.bar(water_means.index, water_means.values, color='turquoise', edgecolor='black')
plt.bar(df["Weight_Change_Category"], df["Water_Intake_Liters"], color='skyblue',edgecolor = "black")
plt.title('Hydration Level by Weight Change Category')
plt.xlabel('Weight Change Category')
plt.ylabel('Water Intake Liters')
plt.show()
```

```python
#17. Are users with better sleep patterns also more active?  Scatter Plot / Heatmap
plt.figure(figsize=(8, 5))
sns.scatterplot(data=df,x='Sleep_Hours', y='Exercise_Minutes', hue='Weight_Change_Category', palette='tab10')
plt.title('Sleep vs Exercise Activity')
plt.xlabel('Sleep Hours')
plt.ylabel('Exercise Minutes')
plt.show()

corr = df[['Sleep_Hours', 'Exercise_Minutes']].corr()
plt.figure(figsize=(6, 5))
sns.heatmap(corr, annot=True, cmap='coolwarm')
plt.title('Correlation Between Sleep, Exercise')
plt.show()
```

```python
#18. Does more exercise correlate with better mood?  Scatter Plot / Line Chart
plt.figure(figsize=(8, 5))
sns.scatterplot( data=df,x='Exercise_Minutes', y='Mood_Score', hue='Weight_Change_Category', palette='Dark2')
plt.title('Exercise vs Mood Score')
plt.xlabel('Exercise Minutes')
plt.ylabel('Mood Score')
plt.tight_layout()
plt.show()


mood_by_exercise = df.groupby('Exercise_Minutes')['Mood_Score'].mean()
plt.figure(figsize=(8, 5))
plt.plot(mood_by_exercise.index,mood_by_exercise.values,color = "green",marker= "o")
plt.title('Mood Score by Exercise Level')
plt.xlabel('Exercise Minutes')
plt.ylabel('Mood Score')
plt.tight_layout()
plt.show()
```

```python
#19. How do calories and exercise minutes together affect weight category?  Scatter Plot (color by category) / Pairplot
plt.figure(figsize=(8, 5))
sns.scatterplot( data=df,x='Calories_Consumed', y='Exercise_Minutes', hue='Weight_Change_Category', palette='tab10')
plt.title('Calories vs Exercise Minutes by Weight Category')
plt.xlabel('Calories Consumed')
plt.ylabel('Exercise Minutes')
plt.tight_layout()
plt.show()

sns.pairplot(df, vars=['Calories_Consumed', 'Exercise_Minutes'], hue='Weight_Change_Category', palette='Set2')
plt.suptitle('Pairplot of calories consumed  & exercise minutes by Weight Category')
plt.show()
```

```python
#20. Explore how protein, carbs, and sugar affect BMI.  Heatmap / Pairplot
corr = df[["Protein_Grams","Carbs_Grams","Sugar_Grams","BMI"]].corr()
plt.figure(figsize=(6, 5))
sns.heatmap(corr, annot=True, cmap='coolwarm')
plt.title('Correlation Between Macronutrients and BMI')
plt.show()

sns.pairplot(df[["Protein_Grams","Carbs_Grams","Sugar_Grams","BMI"]])
plt.suptitle('Pairplot of  Calories Consumed and  Exercise Minutes')
plt.show()
```

```python
#21. Are users with better sleep and mood more likely to maintain weight?  Heatmap / Area Chart
#heatmap_data = df.groupby('Weight_Change_Category')[['Sleep_Hours', 'Mood_Score']].mean().round(2)

avg_metrics = df.groupby('Weight_Change_Category')[['Sleep_Hours', 'Mood_Score']].mean()
plt.figure(figsize=(8, 5))
plt.stackplot(avg_metrics.index,avg_metrics['Sleep_Hours'], avg_metrics['Mood_Score'],labels=['Sleep Hours', 'Mood Score'],alpha=0.7)
plt.title('Sleep & Mood Patterns by Weight Change Category')
plt.xlabel('Weight Change Category')
plt.ylabel('Average Value')
plt.legend(loc='upper left')
plt.tight_layout()
plt.show()
```
